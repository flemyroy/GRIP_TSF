```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Loading the dataset
d=datasets.load_iris()
df=pd.DataFrame(d.data,columns=d.feature_names)
df.head()
```

```python
wcss_list=[]

x = df.iloc[:, [0, 1, 2, 3]].values
```

```python
#Finding optimum number of clusters i.e k-value
#Elbow Method
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1,11),wcss_list)
plt.title("The Elbow Method Graph")
plt.xlabel("No. of Clusters (k)")
plt.ylabel("wcss_list")
#within clusters sum of squares
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_pred=kmeans.fit_predict(x)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x[y_pred ==0,0],x[y_pred == 0,1],s=75,c='red',label="Iris-Setosa")
plt.scatter(x[y_pred ==1,0],x[y_pred == 1,1],s=75,c='darkgreen',label="Iris-Versicolour")
plt.scatter(x[y_pred ==2,0],x[y_pred == 2,1],s=75,c='blue',label="Iris-Virginica")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=250,c='gold',label="Centroid",marker='*')
plt.title("Clusters of Iris")
plt.legend()
plt.show()
```

The Clusters are formed with their respective centroids
